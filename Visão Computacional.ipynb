## Visão computacional
#### Processamento de imagens com OpenCV.

[OpenCV](https://opencv.org/) umas da melhores ferramentas para tarabalhar com imagens, gratuida lançada pela Intel em 1999.

```python
from matplotlib import pyplot as plt
import cv2
%matplotlib inline
```

### Processo de Resize com OpenCV
Lembrando que o RGB no opencv é BGR ou seja invertido.

```python
image = cv2.imread('./Cap02/opencv/images/familia.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Ajustando a largura da imagem

```python
ratio = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * ratio))

resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

Ajustando a altura da imagem

```python
ratio = 50.0 / image.shape[1]
dim = (int(image.shape[1] * ratio), 50)

resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```
