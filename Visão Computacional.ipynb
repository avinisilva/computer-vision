### Manipulando pixels

```python
image = cv2.imread('./Cap02/scikit-image/images/image02.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Pegando o pixel nas codenadas <b>x = 0</b> e <b>y = 0</b>

```python
# opencv grava o RGB de forma inversa para maior eficiência
(b, g, r) = image[0, 0]
print('Pixel in (0, 0) - Red: {}, Green: {}, Blue: {}'.format(r, g, b))
```

Fatiando um grpo de pixels da imagem, pois temos uma matriz e podemos usar o slice do numpy.

```python
corner = image[0:100, 0:100]
plt.imshow(cv2.cvtColor(corner, cv2.COLOR_BGR2RGB))
plt.show()
```

Modificando a cor do grupo de pixels fatiados da imagem

```python
image[0:100, 0:100] = (0, 255, 0)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Criando objetos sobre a imagem

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
for y in range(0, image.shape[0], 50):
    for x in range(0, image.shape[1], 50):
        image[y: y+15, x: x+15] = (0, 255, 255)
        
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Gravando texto sobre a imagem 

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(image, 'Hello guys!', (100,100), font, 2, (255, 255, 255), cv2.LINE_8)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Shift

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
M = np.float32([[1, 0, -50], [0, 1, -90]])
shifted = cv2.warpAffine(image, M, (image.shape[1], image.shape[0]))
plt.imshow(cv2.cvtColor(shifted, cv2.COLOR_BGR2RGB))
plt.show()
```

### Rotacionar a imagem

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
(h, w) = image.shape[:2]
# Encontrar pexel central
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()
```

### Redimencionamento e Aspect Ratio

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Calculado o Aspect Ratio (razão entre largura e altura)
ratio = 150.0 / image.shape[1]
dimension = (150, int(image.shape[0] * ratio))

redized = cv2.resize(image, dimension, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(redized, cv2.COLOR_BGR2RGB))
plt.show()
```

### Flipping

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Operações aritméticas

```python
image = cv2.imread('./Cap02/scikit-image/images/image03.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

### Processo de Resize com OpenCV
Lembrando que o RGB no opencv é BGR ou seja invertido.

```python
image = cv2.imread('./Cap02/pillow/images/image01.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

Ajustando a largura da imagem

```python
ratio = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * ratio))

resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```

Ajustando a altura da imagem

```python
ratio = 50.0 / image.shape[1]
dim = (int(image.shape[1] * ratio), 50)

resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
plt.show()
```
